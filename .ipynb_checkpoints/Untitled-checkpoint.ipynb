```python
import json
import pandas as pd
import matplotlib as plt
```

```python
#load data
file = "all-in_podcast.json"
data = None
with open(file, 'r') as f:
    data = json.load(f)

channel_id, stats = data.popitem()
print(channel_id)
channel_stats = stats["channel_statistics"]
video_stats = stats["video_data"]
```

```python
#Channel Statistics
total_views = int(channel_stats["viewCount"])
total_subscribers = int(channel_stats["subscriberCount"])
total_videos = int(channel_stats["videoCount"])
print(f"Total Views: {total_views:,} \nTotal Subs: {total_subscribers:,} \nTotal Videos: {total_videos:,}")
```

```python
#Video statistics, sort videos by views
sorted_vids = sorted(video_stats.items(), key=lambda item: int(item[1]["viewCount"]), reverse=True)
```

```python
stats = []
for vid in sorted_vids:
    video_id = vid[0]
    title = vid[1]["title"]
    views = int(vid[1]["viewCount"])
    likes = int(vid[1]["likeCount"])
    comments = int(vid[1]["commentCount"])
    stats.append([title,views,likes,comments])
```

```python
#create dataframe

df = pd.DataFrame(stats, columns=["title", "views", "likes", "comments"])
df
```

```python
top10 = df.head(10).sort_values(by="views", ascending=True)
ax = top10.plot.barh(x="title", y="views", figsize=(15,8), fontsize=14)
```

```python
bottom10 = df.tail(10).sort_values(by='views', ascending=True)
ax = bottom10.plot.barh(x="title", y="views", figsize=(12,8), fontsize=14)
```

```python
#Likes vs Dislikes
likes_avg = top10['likes'].mean()
comments_avg = top10['comments'].mean()

print("Average Likes and Comments Top 10 Videos\n")
print(f"Average Likes: {likes_avg:,} \nAverage Comments: {comments_avg:,}")
```

```python
#Likes vs Dislikes For Least Viewed Videos
likes_avg = bottom10['likes'].mean()
comments_avg = bottom10['comments'].mean()

print("Average Likes and Comments Least Viewed 10 Videos\n")
print(f"Average Likes: {likes_avg:,} \nAverage Comments: {comments_avg:,}")
```
